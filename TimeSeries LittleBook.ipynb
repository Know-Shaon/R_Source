```r
kings <- scan("http://robjhyndman.com/tsdldata/misc/kings.dat",skip=3)
kingstimeseries <- ts(kings)
#-----------------------------------------
births <- scan("http://robjhyndman.com/tsdldata/data/nybirths.dat")
birthstimeseries <- ts(births, frequency=12, start=c(1946,1))
#-----------------------------------------
souvenir <- scan("http://robjhyndman.com/tsdldata/data/fancy.dat")
souvenirtimeseries <- ts(souvenir, frequency=12, start=c(1987,1))
#----------------------------------
rain <- scan("http://robjhyndman.com/tsdldata/hurst/precip1.dat",skip=1)
rainseries <- ts(rain,start=c(1813))
#----------------------------------
skirts <- scan("http://robjhyndman.com/tsdldata/roberts/skirts.dat",skip=5)
skirtsseries <- ts(skirts,start=c(1866))
```

```r
#Single Exponential Smoothing
#https://www.itl.nist.gov/div898/handbook/pmc/section4/pmc431.htm
#Double Exponential Smoothing
#https://www.itl.nist.gov/div898/handbook/pmc/section4/pmc433.htm
#Triple Exponential Smoothing
#https://www.itl.nist.gov/div898/handbook/pmc/section4/pmc435.htm
rainseriesforecasts <- HoltWinters(rainseries, beta=FALSE, gamma=FALSE)
```

```r
rainseriesforecasts
```

```r
plot(rainseriesforecasts)
```

```r
rainseriesforecasts$SSE
```

```r
HoltWinters(rainseries, beta=FALSE, gamma=FALSE, l.start=23.56)
```

```r
#install.packages("forecast")
library("forecast")
```

```r
rainseriesforecasts2 <- forecast:::forecast.HoltWinters(rainseriesforecasts, h=8)
```

```r
plot(rainseriesforecasts2)
```

```r
acf(rainseriesforecasts2$residuals[2:length(rainseriesforecasts2$residuals)], lag.max=20)
```

```r
plot.ts(rainseriesforecasts2$residuals)
```

```r
hist(rainseriesforecasts2$residuals[2:length(rainseriesforecasts2$residuals)])
```

```r
skirtsseriesforecasts <- HoltWinters(skirtsseries, gamma=FALSE)
```

```r
skirtsseriesforecasts
```

```r
plot(skirtsseriesforecasts)
```

```r
skirtsseriesforecasts<-HoltWinters(skirtsseries, gamma=FALSE, l.start=608, b.start=9)
```

```r
skirtsseriesforecasts2 <- forecast:::forecast.HoltWinters(skirtsseriesforecasts, h=19)
plot(skirtsseriesforecasts2)
```

```r
acf(skirtsseriesforecasts2$residuals[3:length(skirtsseriesforecasts2$residuals)], lag.max=20)
Box.test(skirtsseriesforecasts2$residuals[3:length(skirtsseriesforecasts2$residuals)], lag=20, type="Ljung-Box")
```

```r
plot.ts(skirtsseriesforecasts2$residuals[3:length(skirtsseriesforecasts2$residuals)])            # make a time plot
plotForecastErrors(skirtsseriesforecasts2$residuals[3:length(skirtsseriesforecasts2$residuals)]) # make a histogram
```
